# File IO

```python
#to make a file pointer, make a variable and set it equal to the open function
#the default mode is read mode
file = open('quotes.txt')

#the file pointer will initially point to the start of the file
#it will seperate the file into strings, separated at new line characters
#once it reads a character 
for line in file:
    print(line)

file.close()
```

```python
#opening a file to write to
#opening a file in write mode erases its contents, before allowiing you write to the file
#if the file does not exist, it will be created
file = open('output.txt', 'w')

for i in range(1, 21):
    file.write('Line number: ' + str(i) + '\n')

file.close()
```

```python
#to write to a file without deleting the contents, you can open it in append mode
file = open('output.txt', 'a')

for i in range(21, 41):
    file.write('Line number: ' + str(i) + '\n')

file.close()
```

## File IO examples

```python
#program that asks a user for an integer, then stores all the words in the scrabble dictionary of that length into a list
#the scrabble dictionary is stored in a file called dictionary.txt, which is in the course folder
size = int(input('Enter an integer: ')) + 1
#As the program reads a line, it will include the newline character
words = []
file = open('dictionary.txt')

for word in file:
    if(len(word) == size):
        words.append(word)
        
file.close()
```

```python
#Now write the list of words to a file called 'words.txt'
file = open('words.txt', 'w')

for word in words:
    file.write(word)
file.close()
```

```python
#A csv file is table of data where the columns are separated by commas, and rows are separated by new lines
#It is often handled in excel, but it can be much more adventageuos to read the file into a python program
#The following file reads in a csv file to store the data in an array
file = open('xyz.csv')
data = []
for line in file:
    data.append([float(x) for x in line.split(sep = ',')])
for x in data:
    print(*x)
file.close()
```
